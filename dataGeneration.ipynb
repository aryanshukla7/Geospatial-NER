```python
from keytotext import pipeline
import cv2
# nlp1 = pipeline("k2t")
nlp2 = pipeline("k2t-base")
nlp3 = pipeline("mrm8488/t5-base-finetuned-common_gen")
```

```python
import pandas as pd
import numpy as np
```

```python
cities_df = pd.read_csv("./data/cities.csv")
states_df = pd.read_csv("./data/states.csv")
countries_df = pd.read_csv("./data/countries.csv")
```

```python
cities_df
```

```python
n_cities = len(cities_df)
n_states = len(states_df)
n_countries =len(countries_df)

df = pd.DataFrame()
sentence_li = []
input_li = []
for i in range(1000):
    city_names = []
    state_names = []
    country_names = []

    input_dict = {}

    for j in range(7):
        x = np.random.randint(4)
        if (x==1):
            city_names.append(cities_df.iloc[np.random.randint(n_cities)]["name"])
            input_dict[city_names[-1]]="city"
        elif (x==2):
            state_names.append(states_df.iloc[np.random.randint(n_states)]["name"])
            input_dict[state_names[-1]]="state"
        elif (x==3):
            country_names.append(countries_df.iloc[np.random.randint(n_countries)]["name"])
            input_dict[country_names[-1]]="country"

    sentence_li.append(nlp3([city_names + state_names + country_names]))
    input_li.append(input_dict)
df = pd.DataFrame({"Sentences":sentence_li, "Places":input_li})
df.to_csv("t5_1_1000.csv")
```
